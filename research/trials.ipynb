```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
project_spec = '''
For this assignment in the Advanced Data Engineering 2025 course, students must design and implement a complete data ingestion, processing, and analytics pipeline for a fictional online retail company that wants to better understand customer purchasing behaviour and identify trends across product categories. The project requires students to collect at least three complementary datasets: a transaction dataset containing individual customer purchases with timestamps, product identifiers, and price information; a product metadata dataset including category, subcategory, supplier, and basic descriptive fields; and a customer dataset containing anonymised demographic attributes such as age range, region, and loyalty status. Students must clean and normalise all datasets, resolve missing values, and ensure that keys across tables are consistent. They must also implement at least two forms of feature engineering, such as computing rolling averages of purchase frequency, extracting time-based features like weekend versus weekday activity, or inferring product category popularity metrics. After preparing the data, students are required to design a scalable data pipeline using Python and SQL, with optional use of distributed technologies such as Spark or Dask. The pipeline must ingest raw CSV or JSON files, store them in a structured relational database of the student’s choice, transform them using a reproducible ETL or ELT workflow, and finally produce analytical outputs that can be queried through a single consolidated table or materialised view. Students must produce at least three analytical deliverables: one descriptive summary of customer purchasing patterns, one diagnostic analysis identifying which product categories generate the highest revenue, and one predictive component where they build a simple model, such as logistic regression or gradient boosting, to estimate the likelihood that a customer will make a repeat purchase within 30 days. Model performance should be evaluated using appropriate metrics such as accuracy, AUC, or F1 score, and students must justify their modelling choices. All code must follow a clear directory structure with separate folders for raw data, processed data, SQL scripts, ETL scripts, and analysis notebooks. A configuration file must specify database credentials and pipeline parameters without hard-coding sensitive details. Students must also provide a short command-line interface or script that reruns the entire pipeline from ingestion to final analysis. A written report of no more than 5 pages must explain the datasets, cleaning steps, pipeline design, engineered features, database schema, analytical findings, model evaluation, limitations, and possible extensions. Submissions must include a PDF report, access to a private GitHub repository containing all code, an exported database schema, and a brief README explaining how to run the pipeline end-to-end. The project is worth 50 percent of the overall course grade and is due on Friday 28 February 2025 at 17:00 UK time. Late submissions follow the standard School of Informatics penalty policy, and group work is not permitted. Assessment will focus on data quality and preprocessing, pipeline correctness and reproducibility, modelling and analysis quality, clarity of the report, and the structure and readability of the submitted code.
'''
```

```python
model_prompt = f'''
Extract all key information from the following project specification. Return a json style output with using the points below as keys. Use bullet points for every item.

Deadline

Weighting (percentage of the course grade)

Short Description (2–3 sentences summarising the project)

Deliverables (bullet list of what the student must submit)

Project Breakdown

Break the project into its major sections or components

For each section, provide a concise summary of 2–4 sentences

Important Notes

Include information such as extension policies, group work rules, late penalties, required technologies, reporting requirements, formatting constraints, or any other emphasised details in the specification

Text to analyse:
{project_spec}

Make sure the answer is clean, structured, and readable. Do not omit any key detail and do not infer key facts that are not present.
'''
```

```python
load_dotenv()
client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=os.environ["HF_TOKEN"],
)

completion = client.chat.completions.create(
    model="MiniMaxAI/MiniMax-M2:novita",
    messages=[
        {
            "role": "user",
            "content": model_prompt
        }
    ],
)

print(completion.choices[0].message.content)
```
